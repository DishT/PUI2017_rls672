Generate 100 2-dimensional random datapoints in a Gaussian distribution. Call the array ReprRand: 

```python
ReprRand = np.random.randn(2,100)
```

Generate 50 more random 2x100 numpy arrays. Store them in ReprRandAll, all with different mean. 

```python
ReprRandAll = np.zeros((50, 2, 100))
```

Now let's see what ReprRandAll looks like 

```python
ReprRandAll
```

And check the shape of ReprRandAll

```python
ReprRandAll.shape
```

```python
for i in range(50): 
    mu = np.random.randn(2,100)
    sigma = np.random.randn(2,100)
    ReprRandAll[i] = sigma * ReprRand + mu
```

Import pylab and give command to print in jupyter

```python
import pylab as pl
%pylab inline
```

Now let's plot

```python
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Plot')

for i in range(50):
    for j in range(100):
        pl.plot(ReprRandAll[i][0][j], ReprRandAll[i][1][j], 'o')
```
